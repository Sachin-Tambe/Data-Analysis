##  Does more time on site lead to bigger orders?

```python
df['Total Order Value'] = pd.to_numeric(df['Total Order Value'] , errors='coerce')
average_order_value =  df['Total Order Value'].mean() 
df['Above Average Order'] = df['Total Order Value'] > average_order_value
```

```python
sns.scatterplot(x=df['Total Order Value'] , y=df['Time Spent on Website'] , hue=df['Above Average Order'] )
plt.show()
```

####  There’s no clear link between how long someone stays and how much they spend.

##  How many orders do we get each month?

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d-%m-%Y")
```

```python
order_count_by_month = df['Order Date'].dt.to_period('M').value_counts()
```

```python
plt.figure(figsize=(11, 11))
plt.pie(order_count_by_month , labels=order_count_by_month.index , autopct='%1.1f%%', wedgeprops={'edgecolor': 'white'})
plt.title('Order Distribution by Month')
plt.tight_layout()
plt.show()
```

## How does our revenue change month-to-month?

```python
df['Year-Month'] = pd.to_datetime(df['Year'].astype(str) + " " + df['Month'] , format='%Y %B' )
```

```python
monthly_revenue = df.groupby(df['Year-Month'])['Total Order Value'].sum().reset_index()
```

```python
plt.figure(figsize=(10,6))
plt.plot(monthly_revenue['Year-Month'] , monthly_revenue['Total Order Value'] , marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Revenue is increasing month by month which is good sign for the organization
